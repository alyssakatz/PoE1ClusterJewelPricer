```python
import requests
import json
import pandas as pd
```

```python
request = requests.get('https://poe.ninja/poe1/api/data/index-state').text
snapshots = json.loads(request)
snapshots = pd.json_normalize(snapshots['snapshotVersions']).apply(pd.Series)
snapshots
```

```python
pd.options.display.float_format = '{:.2f}'.format

league = 'Mercenaries'
category = 'UniqueJewel'
items = ['Brute Force Solution', 'Energy From Within', 'Fertile Mind', 'Fevered Mind', 'Fortress Covenant', 'Healthy Mind', 'Powerlessness', 'Reckless Defence', 'Replica Powerlessness', 'To Dust']
```

```python
uri = "https://poe.ninja/api/data/itemoverview?league=" + league + "&type=" + category

def get_item_overviews(league, category):
    df = pd.read_json(uri)
    df = df['lines'].apply(pd.Series)
    df = df[df['name'].isin(items)]
    df = df.reset_index(drop=True)
    return df

overviews = get_item_overviews(league, category)
overviews
```

```python
item_uri_base = 'https://poe.ninja/poe1/api/economy/stash/current/item/history?league=' + league + '&type=' + category + '&id='

def calculate_weekly_stats(id):
    item_uri = item_uri_base + id
    tmp = pd.read_json(item_uri)
    tmp = tmp.reset_index(names='daysSinceStart')
    tmp = tmp.rename(columns={'value': 'chaos'})
    tmp = tmp.drop(columns=['daysAgo'])
    tmp['weekNum'] = tmp['daysSinceStart'] // 7

    stats = tmp.groupby('weekNum').agg(
        chaos_avg=('chaos', 'mean'))

    
#    stats = tmp.groupby('weekNum').agg(
#        count_avg=('count', 'mean'),
#        count_std=('count', 'std'),
#        chaos_avg=('chaos', 'mean'),
#        chaos_std=('chaos', 'std'))

    stats = stats.T
    flattened = pd.DataFrame()

    for col in stats:
        week_number = col
        week_prefix = f'week{week_number}'
        for index, value in stats[col].items():
            new_column_name = f'{week_prefix}_{index}'
            flattened[new_column_name] = [value]
    
    return flattened

def calculate_item_prices(df):
    item_prices = df['id'].astype(str).apply(calculate_weekly_stats)
    item_prices = pd.concat(list(item_prices), ignore_index=True)
    item_prices = df[['id', 'name', 'baseType', 'chaosValue']].join(item_prices)
    item_prices.rename(columns={'chaosValue': 'avg_chaos'})
    return item_prices

item_prices = calculate_item_prices(overviews)
item_prices
```

```python
import re

wiki_uri = 'https://www.poewiki.net/wiki/'
corruption_chance_pattern = r"(?:yielded )(\d+)"

def get_corruption_chance(name):
    uri = wiki_uri + name
    html = requests.get(uri).text
    match = re.search(corruption_chance_pattern, html)

    if match:
        return int(match.group(1)) / 576 * 100
    else:
        return 'Unknown'

category = 'UniqueJewel'
items = ["Brute Force Solution", "Energy From Within", "Fertile Mind", "Fevered Mind", "Fortress Covenant", "Healthy Mind", "Powerlessness", "Reckless Defence", "To Dust", "Atziri's Reign", "Efficient Training", "Energised Armour", "Fragility", "Inertia", "Rain of Splinters", "Careful Planning", "Chill of Corruption", "Fluid Motion", "Hidden Potential", "Pacifism", "Quickening Covenant", "Seething Fury", "Self-Flagellation", "The Golden Rule"]
corrupted_jewel_prices = calculate_item_prices(get_item_overviews(category, items))
url_names = corrupted_jewel_prices['name'].str.replace(' ', '_')
url_names = url_names.str.replace("'", '%27')
values = url_names.apply(get_corruption_chance)
corrupted_jewel_prices.insert(loc=3, column='corruptChance', value=values)
corrupted_jewel_prices = corrupted_jewel_prices.sort_values(by=['baseType', 'corruptChance', 'name'], ascending=[True, False, True])
corrupted_jewel_prices
```
